```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_dataset_1=pd.read_csv('..//dataset//final_dataset.csv')
df_dataset_2=pd.read_excel('..//dataset//Dataset_IoV.xlsx')
```

```python
print(df_dataset_1.shape, df_dataset_2.shape)  # Number of rows and columns
print(df_dataset_1.info(), df_dataset_2.info())  # Data types and non-null counts
print(df_dataset_1.head(), df_dataset_2.head())  # First few rows
```

```python
print(df_dataset_1.isnull().sum(), df_dataset_2.isnull().sum())  # Count of missing values
```

```python
print(df_dataset_1.describe(), df_dataset_2.describe())  # Summary statistics
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.heatmap(df_dataset_1.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix - Dataset 1")
plt.show()

plt.figure(figsize=(10,6))
sns.heatmap(df_dataset_2.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix - Dataset 2")
plt.show()
```

```python
df_dataset_1.boxplot(figsize=(10,6))
plt.title("Boxplot - Dataset 1")
plt.show()

df_dataset_2.boxplot(figsize=(10,6))
plt.title("Boxplot - Dataset 2")
plt.show()
```

```python
df_dataset_1.hist(figsize=(12,8), bins=30)
plt.suptitle("Feature Distribution - Dataset 1")
plt.show()

df_dataset_2.hist(figsize=(12,8), bins=30)
plt.suptitle("Feature Distribution - Dataset 2")
plt.show()
```
